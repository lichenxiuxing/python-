```python
import matplotlib.pyplot as plt
import numpy as np
# 第一部分 pyplot的使用
# plot的作用是绘图
plt.plot([1,3,5,6])
# 添加纵坐标
plt.ylabel('some numbers')
# show用来显示图片
plt.show()
```

```python
#默认是蓝色 直线
plt.plot([1,3,5,6],[3,5,3,3],'g:')
# 设置坐标轴
plt.axis([1,8,1,12])
```

```python
# 绘制多条线
a=np.arange(0.,5.,0.5)
plt.plot(a,2*a,a,a**2,a,a**3)
```

```python
# 第二部分 figure
# figure 的作用是开辟一个画板，常用参数有id figsize facecolor
plt.figure('one',facecolor="red")
plt.plot([2,5,5])
# 一个画板上画多个图
plt.figure('2')
plt.subplot(2,1,1)
plt.plot([3,5,1],label='label1')
plt.xlabel('year',fontsize=14)
plt.ylabel('name')
# 添加标题
plt.title('this is a title')
# 添加文本注释
plt.text(2,3,'look at me')
plt.legend()

plt.subplot(2,1,2)
plt.plot([6,6,3],label='label2')
plt.show
```

```python
# 第三部分 plt文本相关函数使用
# 一般直接使用plot在axes上绘制，也可以直接获取axes进行绘图。
# 手动创建一个figure
f1=plt.figure('one')
# 获取当前axes
current_axes=plt.gca()
```

```python
# 第四部分 中文乱码的问题
#设置这个就可以了，
plt.rcParams['font.sans-serif']=['SimHei'] #用来正常显示中文标签 
plt.rcParams['axes.unicode_minus']=False #用来正常显示负号
plt.figure('今天')
plt.title('明天')
plt.xlabel('昨天')
plt.ylabel('后天')
```

```python
# 第五部分 柱状图
# 通过plt.bar来实现,其中，left,height,bottom,和width（默认0.8）很重要
#plt.bar(left=[10,11],height=[10,15])
# 四个点的坐标为：图展示的时候，默认是居中的
#左下(left-width/2,bottom)，右下(left+width/2,bottom)，
#左上(left-width/2,bottom+height)，右上(left+width/2,bottom+height)
#即(9.6,0)，(10.4,0)，(9.6,10)，(10.4,10)
plt.bar([10,11],[10,15])
# 显示网格线
plt.grid(True)
plt.show
```

```python
# 将width改为1,看起来会更为方便
plt.bar([10,12],[10,15],width=1)
```

```python
# 还可以修改对齐方式
plt.bar([10,12],[10,15],width=1,align='edge')
```

```python
# 对每一个柱状图都可以进行将width和bottom设置
plt.bar([10,12],[10,15],width=[0.5,1],bottom=(1,-1))
```

```python
# 设置填充颜色和边框颜色
plt.bar([10,12,14],[10,15,20],facecolor='red',edgecolor='yellow')
```

```python
# 设置不同的颜色
plt.bar([10,13,15],[10,15,16],color=['red','blue','yellow'])
```

```python
# 边框的样式、宽度以及填充、标签
plt.bar([12,15,16],[10,12,15],facecolor='blue',edgecolor='yellow',linestyle='--',linewidth=3,hatch='-',tick_label=['一','二','三'])
```

```python
# 堆叠柱状图与并列柱状图，都是计算好位置就可以
plt.bar([10,14,12],[13,15,10],bottom=0,color='red')
plt.bar([10,14,12],[2,5,6],bottom=[13,15,10],color='blue')
```

```python
plt.bar([10,14,12],[13,15,10],width=1,color='red')
plt.bar([11,15,13],[13,15,10],width=1,color='blue')
```

```python
# 6 水平条形图，默认的height是0.8
data=np.random.randint(10,100,size=10)
print(data)
print(np.arange(data.size))
plt.barh(np.arange(data.size),data,label='蓝色')
plt.legend()

```

```python
data=np.random.randint(10,100,size=10)
plt.barh(np.arange(data.size),data,label='蓝色')
plt.legend()
# 设置坐标轴的长度及标签
plt.xticks(np.arange(0,100,10))
plt.yticks(np.arange(data.size),('的','数','据','反','馈','的','时','间','肯','是'))
```

```python
# 对照图 将X轴变为负的
data=np.random.randint(10,100,size=10)
plt.barh(np.arange(data.size),data,label='蓝色')
plt.barh(np.arange(data.size),-data,label='不确定')
plt.legend()
```

```python
# 第七部分 折线图和曲线图
# 最开始掌握的用plot绘制的就是折线图
x=np.arange('2019-04-01','2019-04-30',dtype=np.datetime64)
y=np.random.randint(0,100,size=29)
y1=np.random.randint(0,100,size=29)
plt.plot(x,y,color='red',label='pc')
plt.plot(x,y1,color='blue',label='mvp')
plt.title('有点意思')
plt.xlabel('日程')
plt.ylabel('登录人数')
# 遍历每一个点，为其加以text
for i,y in list(zip(x,y)):
    plt.text(i,y+1,y,fontsize=12)
plt.legend()
```

```python
x=np.linspace(-2*np.pi,2*np.pi,1000)
y1,y2=np.sin(x),np.cos(x)
plt.plot(x,y1,color='blue',label='又去')
plt.plot(x,y2,color='yellow',label='第三节课')

plt.legend()
```

```python
# 散点图
N=100
x=np.random.rand(N)
y=np.random.rand(N)
plt.scatter(x,y,color='blue',label='df')

plt.legend()
```
